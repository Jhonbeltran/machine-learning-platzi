## Ahora solucionaremos el problema de los datos faltantes (nulos o NaN). 

<div class="alert alert-success">
  Los datos faltantes generan problemas con muchos algoritmos de ML. Es por esto que existen distintas estrategias para lidiar con ellos.
</div>

```python
help(pd.Series.value_counts)
```

```python
movies_v2.notnull().apply(pd.Series.value_counts)
```

```python
(movies_v2 != 0).apply(pd.Series.value_counts)
```

```python
available = ((movies_v2 != 0) & (movies_v2.notnull()))
```

```python
available.all(axis=1).value_counts() #Para saber cuantas lineas tienen valores nulos o vacios
```

No podemos entrenar nuestro algoritmo con datos cuya variable objetivo no esta definida o sea nula (valor falso). Eliminemos esas líneas.

```python
mask = available['worldwide_gross']
```

```python
movies_v2 = movies_v2[mask] #Guardamos los datos que si estan completos
```

```python
((movies_v2 != 0) & (movies_v2.notnull())).worldwide_gross.value_counts() 
#Revisamos que la columna no tenga valores faltantes (worldwide_gross)
```

<div class="alert alert-success">
  En el caso de las features que no son la variable objetivo una mejor solución para lidiar con los datos faltantes es <strong>remplazar estos datos por otros que sean manejables</strong> y no afecten la calidad de las predicciones. La estrategia más comun es utilizar la media de todos los ejemplos para la feature dada.
</div>

```python
movies_v2 = movies_v2.drop('movie_title',axis=1) 
#Eliminamos esta columna debido a que necesitamos sólo datos numéricos
```

```python
movies_v2 = movies_v2.drop('duration',axis=1)
#Eliminiamos esta columna debido a que está duplicada
```

```python
movies_v2 = movies_v2.drop('gross',axis=1)
# A esta le faltan muchos datos
```

```python
movies_v2.head(2)
```

```python
movies_v2 = movies_v2[available.screens] #Dejamos sólamente los que no tienen null en la feature screens
```

```python
len(movies_v2)
```

```python
((movies_v2 != 0) & (movies_v2.notnull())).apply(pd.Series.value_counts)
```

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values=np.nan, strategy='mean', axis=1) 
#Esta instancia nos va a ayudar a completar los datos faltantes de las columnas que aun no hemos reparado
```

```python
values = imputer.fit_transform(movies_v2) #Va a quedar de tipo array de numpy
X = pd.DataFrame(values) #Llevamos los datos de nuevo a un dataframe
X.columns = movies_v2.columns # Agregamos las columnas al data Frame
X.index = movies_v2.index # Agregamos los index al data Frame
X.head(2)
```

```python
len(X)
```

```python
movies_v2.values
```

```python
values
```

```python
((X != 0) & (X.notnull())).apply(pd.Series.value_counts)
```

```python
X.to_csv('intermediate_results/X.csv',index=False)
```
